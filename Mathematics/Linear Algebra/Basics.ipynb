```python
import numpy as np
```

```python
# Distance from Origin
A = np.array([2,4])
distance = np.linalg.norm(A)
distance
```

```python
# Eucledian Distance
A = np.array([2,3,4])
B = np.array([5,6,7])

print("vector A: ", A)
print("vector B: ", B)
print("Eucledian Distance: sqrt(",)
difference = A-B
distance = np.linalg.norm(difference)
print(distance)
```

```python
# Scaler Addition & Scaler Substraction
A = np.array([2,3,4])
n = int(input("Enter a scaler to add to vector: "))
print("Original Vector: ", A)

print("After Scaler Addition A+n: ",A+n)
print("After Scaler Substraction A-n: ",A-n)
```

```python
# Dot Product
A = np.array([2,3,4])
B = np.array([5,6,7])
np.dot(A,B)
```
