```python
# importing Dataset -> Sharma-Kohli

url = "https://drive.google.com/uc?export=download&id=1vI6QgVkD4UZHlTEknKDHJGkdwdbmMx__"
batsman = pd.read_csv(url)
batsman
```

```python
# x-> categorical y->numerical
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple Plot
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# label title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Sharma VS Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Color and line (width and size) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color = "green",linestyle = "dashed",linewidth = 2,marker = 'x')
plt.plot(batsman['index'],batsman['RG Sharma'],color = "black",linestyle = "dashdot",linewidth = 3,marker = 'o',markersize = 10)
plt.title('Sharma VS Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend
plt.plot(batsman['index'],batsman['V Kohli'],color = "green",linestyle = "dashed",linewidth = 2,marker = 'x',label = "Kohli")
plt.plot(batsman['index'],batsman['RG Sharma'],color = "black",linestyle = "dashdot",linewidth = 3,marker = 'o',markersize = 10,label = "Rohit")
plt.title('Sharma VS Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend() # prints only when lable parameter is set in the plt.plot func.
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color = "green",linestyle = "dashed",linewidth = 2,marker = 'x')
plt.plot(batsman['index'],batsman['RG Sharma'],color = "black",linestyle = "dashdot",linewidth = 3,marker = 'o',markersize = 10)
plt.title('Sharma VS Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```
