```python
# importing Dataset -> Sharma-Kohli

url = "https://drive.google.com/uc?export=download&id=1vI6QgVkD4UZHlTEknKDHJGkdwdbmMx__"
batsman = pd.read_csv(url)
batsman
```

```python
# x-> categorical y->numerical
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple Plot
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# label title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Sharma VS Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Color and line (width and size) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color = "green",linestyle = "dashed",linewidth = 2,marker = 'x')
plt.plot(batsman['index'],batsman['RG Sharma'],color = "black",linestyle = "dashdot",linewidth = 3,marker = 'o',markersize = 10)
plt.title('Sharma VS Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend
plt.plot(batsman['index'],batsman['V Kohli'],color = "green",linestyle = "dashed",linewidth = 2,marker = 'x',label = "Kohli")
plt.plot(batsman['index'],batsman['RG Sharma'],color = "black",linestyle = "dashdot",linewidth = 3,marker = 'o',markersize = 10,label = "Rohit")
plt.title('Sharma VS Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend() # prints only when lable parameter is set in the plt.plot func.
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color = "green",linestyle = "dashed",linewidth = 2,marker = 'x')
plt.plot(batsman['index'],batsman['RG Sharma'],color = "black",linestyle = "dashdot",linewidth = 3,marker = 'o',markersize = 10)
plt.title('Sharma VS Kohli')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

## Bar Chart

Note:
- Bivariate Analysis
- Numerical Vs Categorical
- Use Case: Aggregate Analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# Horizontal Bar Charts
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.barh(colors,children,color='black')
```

```python
# import dataset
url = "https://drive.google.com/uc?export=download&id=10p6RZu-q5PGXLwsXAfemJbjAIN_A0BGq"
df = pd.read_csv(url)
df
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
# Not a good way
```

```python
plt.bar(np.arange(df.shape[0])-0.2, df['2015'], width = 0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2)
# The numbers preseant at each index/possition its a center position of that bar

plt.xticks(np.arange(df.shape[0]), df['batsman'], rotation = 'vertical')
plt.show()
```

## Stacked Bar Chart

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

## Pie Chart

Note:
- Univariate and Bivariate
- numerical & Categorial
- Use Case: To find contribution on a standard scale

```python
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```
