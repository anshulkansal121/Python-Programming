```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Coloured Scatter plot

```python
iris = sns.load_dataset("iris")
iris.sample(5)
```

```python
iris['species'] = iris['species'].replace({'setosa': 1,'virginica':2,'versicolor':3})
```

```python
# Now to get the relation between sepal_length and petal_length
# parameters: alpha -> controles the opacity of the plot eg: 0.7
# cmap -> defines the color pallette of the scatter plot
# c = different between different clusters
plt.scatter(iris['sepal_length'],iris['petal_length'],c=iris['species'],cmap = "winter")
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# import dataset
url = "https://drive.google.com/uc?export=download&id=1kJ1v2GDxa5uADs3uproeMp588RtRggJ_"
df = pd.read_csv(url)
df = df.head(25)
df
```

```python
# plotting scatter plot between  avg and strike_rate where each point defined by the runs made by that batsman
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s = df['runs'])
# Now let's add annotation for each point
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

## Horizontal and Vertical Lines - Scatter Plot

```python
# let's say i need batsman you have a strike rate greater than 130
# plotting scatter plot between  avg and strike_rate where each point defined by the runs made by that batsman
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s = df['runs'])
# Now let's add annotation for each point
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])

plt.axhline(130,color = 'red')
plt.axvline(30, color = 'red')
```
