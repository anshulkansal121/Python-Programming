```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Coloured Scatter plot

```python
iris = sns.load_dataset("iris")
iris.sample(5)
```

```python
iris['species'] = iris['species'].replace({'setosa': 1,'virginica':2,'versicolor':3})
```

```python
# Now to get the relation between sepal_length and petal_length
# parameters: alpha -> controles the opacity of the plot eg: 0.7
# cmap -> defines the color pallette of the scatter plot
# c = different between different clusters
plt.scatter(iris['sepal_length'],iris['petal_length'],c=iris['species'],cmap = "winter")
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# import dataset
url = "https://drive.google.com/uc?export=download&id=1kJ1v2GDxa5uADs3uproeMp588RtRggJ_"
df = pd.read_csv(url)
df = df.head(25)
df
```

```python
# plotting scatter plot between  avg and strike_rate where each point defined by the runs made by that batsman
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s = df['runs'])
# Now let's add annotation for each point
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

## Horizontal and Vertical Lines - Scatter Plot

```python
# let's say i need batsman you have a strike rate greater than 130
# plotting scatter plot between  avg and strike_rate where each point defined by the runs made by that batsman
plt.figure(figsize=(18,10))
plt.scatter(df['avg'],df['strike_rate'],s = df['runs'])
# Now let's add annotation for each point
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])

plt.axhline(130,color = 'red')
plt.axvline(30, color = 'red')
```

## SubPlots -> Important

- fig - access the total subplots
- ax - list of all the x any y axis plots
- recreated methods - set_title()
- Locating graphs - ax[1, 0]
- Axis Sharing - sharex parameter in subplot method
- nrows and ncols
- add subplots - add_subplot() method

```python
fig,ax = plt.subplots(nrows = 1, ncols = 2,figsize = (15,6))
ax[0].scatter(df['avg'],df['strike_rate'])
ax[1].scatter(df['avg'],df['runs'])
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10, 7)) # figsize -> Total size

ax[0,1].scatter(df['avg'],df['runs'])
ax[1,0].hist(df['avg'])
ax[1,1].hist(df['runs'])

plt.show()
```

## Adding Plots - add_subplots(nrows, ncols, plot_no)

```python
fig = plt.figure(figsize = (15, 8))

ax1 = fig.add_subplot(2, 2, 1)
ax1.scatter(df['avg'],df['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(df['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(df['avg'])

fig.show()
```

### 3D scatter Plot

```python
fig = plt.figure(figsize = (10, 10))
ax = plt.subplot(projection='3d')

ax.scatter3D(df['runs'],df['avg'],df['strike_rate'],marker='o', s = 30, alpha = 0.5)
ax.set_title('IPL batsman analysis')

ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('SR')
```

## 3d Line Plot

```python
x = [0,1,5,25, 36, 47, 58]
y = [0,10,13,20, 35, 44, 53]
z = [0,13,20,39, 41, 50 ,68]

fig = plt.figure()

ax = plt.subplot(projection='3d')
ax.plot3D(x,y,z,color='black', marker = 'o', markersize = 10)
```

## 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p) # Have to provide referance of the plot
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Contour Plots

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
