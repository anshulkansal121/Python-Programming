### DataFrame
`DataFrame` is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object. Like Series, DataFrame accepts many different kinds of input:

* Dict of 1D ndarrays, lists, dicts, or Series
* 2-D numpy.ndarray
* Structured or record ndarray
* A Series
* Another DataFrame

Along with the data, you can optionally pass **index** (row labels) and **columns** (column labels) arguments. If you pass an index and / or columns, you are guaranteeing the index and / or columns of the resulting DataFrame. Thus, a dict of Series plus a specific index will discard all data not matching up to the passed index.

If axis labels are not passed, they will be constructed from the input data based on common sense rules.



````python
# Creation DataFrame from Scratch
# from 2d lists
students = [
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,50,2]
]
pd.DataFrame(students,columns=['iq','marks','package'])
````

````python
# From dict
student_dict = {
    'name':['Anshul','Aakash','Yash','Vinayak'],
    'iq':[100,90,120,80],
    'marks':[80,70,100,50],
    'package':[10,7,14,2]
}
students = pd.DataFrame(student_dict)
students.set_index('name',inplace = True)
students
````

### Fetching Data From DataFrame - Column Wise (Slicing)
Selecting columns from a Dataframe can be done in two ways:
1. Single Column
   * Syntax: ```dataframename[column_name]```
     
   **Note** : Dtype of the returned object is ```Series``` here
2. Multiple Column:
   * Syntax: ```dataframe[[list of column seperated column name]]```
     
   **Note** : Dtype of the returned object is ```DataFrame``` here.

````python
# Single Column
movies['actors']
ipl['Venue']
````

````python
# Multiple Column
# Use fancy indexing - pass a list of column name as index
ipl[['Team1','Team2','WinningTeam']]
````

### Fetching Data From DataFrame - Row Wise (Slicing)
This can be done using:
* **iloc**: searches using index position
* **loc**: searched using index location

Each of above defined method can be used to fetch either single row or multiple rows.

**NOTE** : `iloc` is avalibale all the time whereas `loc` can be used only when we have index values other than integer nums

````python
# iloc - Fetching single row
movies.iloc[4]
````

````python
# iloc - Fetching multiple rows
movies.iloc[1:5]
````

````python
# iloc - Fancy indexing
movies.iloc[[0,4,5,6]]
````

````python
# loc - Single Row
students.loc['Anshul']
````

````python
# loc - Fancy indexing
students.loc[['Anshul','Aakash','Vinayak']]
````

````python
# loc - Fetching multiple rows
students.loc['Anshul':'Yash']
````

### Fetching both Columns and rows simultaneously


````python
# Statement: I need to fetch first 4 movies and their title, original title and runtime
# using iloc we can only use Index not labels of index.
movies.iloc[0:4,[0,5,8]]
````

````python
# Same can be done using loc as follow
movies.loc[0:4,['title_x','original_title','runtime']]
````

### Filtering DataFrame - Important
For this exercise we will be learning this through a series a Question and will apply the previous concepts.

````python
# DataFram we are using is IPL
ipl.head(5)
````

````python
# Find all the final winners
# Approach: we have a column name MatchNumer that say which match it is and obvio every season have only 1 final match
# Firstly fetch the final match and then select the Season and winning team out of it

ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']].sort_values('Season')
````

````python
# How many Super Over Finishes have occured
# Approach : We have a SuperOver Column with Y for Yes and N for N so we will fetch the rows that have SuperOver as Y and then count all its coourences

ipl[ipl['SuperOver']=='Y'].shape[0]
````

````python
# How many matches has CSK won in Kolkata
ipl[(ipl['City']=='Kolkata') & (ipl['WinningTeam']=='Chennai Super Kings')].shape[0]
````

````python
# Toss winner is match Winner in Percentage
ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0] * 100
````

````python
# Movies with rating greater than 8 and votes > 10000
movies[(movies['imdb_rating'] > 8 ) & (movies['imdb_votes'] > 10000)]
````

````python
# Action Movies with rating higher than 7.5
movies[(movies['genres'].str.contains('Action')) & (movies['imdb_rating'] > 7.5)]
````
