## 1. pd.read_csv()

- Local Disk / in the same folder
- From a URL
- Parameters
  - sep -> defines the seperator of the file eg. tsv -> '\t'
  - index_col -> creates an index out of one of the column from the dataset
  - header -> defines the header from the data itself
  - usecols -> defines the col that we need from the data
  - skiprows -> defines the rows that you want to skip while loading dataset from csv
  - nrows -> defines the number of rows you need to import from the dataset
  - encoding ->
  - dtype -> defines/changes the datatype of a column(s) at the time of importing
  - parse_dates -> list of columns having datetime format/dtype
  - na_values -> used to define / replace the null/nan values
  - chunksize -> when dealing with huge dataset we can load data in chunks
    

```python
# Local Disk
filename = ""
pd.read_csv(filename)
```

```python
# From URL
# parameter sep -> used to define the seperator of the file
pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv",sep = "\t")
```

```python
# parameter: sep
pd.read_csv('Data/movie_titles_metadata.tsv',sep = '\t')
```

```python
# in the above imported data we don't have a header so by default out row becomes the header
pd.read_csv('Data/movie_titles_metadata.tsv',sep = '\t',names = ['s.no','name','release_year','rating','votes','genre'])
```

```python
pd.read_csv('Data/aug_train.csv')
# here we have index col and also enrolle_id so making it as index
pd.read_csv('Data/aug_train.csv',index_col='enrollee_id')
```

```python
pd.read_csv("Data/test.csv")
# here we have header as first row of the data
pd.read_csv("Data/test.csv",header= 1)
```

```python
pd.read_csv("Data/aug_train.csv", usecols = ['enrollee_id','city','gender'])
```

```python
pd.read_csv("Data/aug_train.csv",skiprows = [1,2], nrows=100)
```

```python
encoding_formats = [
    'UTF-8',  # 8-bit Unicode Transformation Format
    'UTF-16',  # 16-bit Unicode Transformation Format
    'UTF-32',  # 32-bit Unicode Transformation Format
    'Latin-1'  # Another name for ISO-8859-1
    'ISO-8859-1',  # Latin-1, single-byte encoding for Western European languages
    'ASCII',  # American Standard Code for Information Interchange
    'Windows-1252',  # A superset of ISO-8859-1 used in Windows
]
pd.read_csv("Data/zomato.csv",encoding='latin-1').head()
```

## to_csv

```python
df = pd.read_csv("Data/deliveries.csv")
df
```

```python
# New Dataset -> Batsmen name and run he scored in total
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_csv('batsman_runs.csv',index=False)
```
