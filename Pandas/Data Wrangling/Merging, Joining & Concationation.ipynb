### loading multiple dataframe for practice in this exercise

```python
courses = pd.read_csv("https://docs.google.com/spreadsheets/d/1ogR_J3P1eNQ_1joPhnn4Rb5bj8QD0o_ivPpLpAhhcA0/export?format=csv&gid=1981096800")
students = pd.read_csv("https://docs.google.com/spreadsheets/d/1-fRhX784kdcCx6faoiYazM5w3jVWYzzeaPLiYz4tQiQ/export?format=csv")
dec = pd.read_csv("https://docs.google.com/spreadsheets/d/1HnMXExLHTb-gfwJjLVM6YtmdSn8V0BKEpvwzg-LG9Y8/export?format=csv&gid=621189259")
nov = pd.read_csv("https://drive.google.com/uc?export=download&id=1jFUPpWwUcT6x8I6V4xN2naKKO9eK_bjE")

matches = pd.read_csv("https://drive.google.com/uc?export=download&id=1IRJJ-mucrtgeOcveCQF-a8CgwQBDJLOl")
delivery = pd.read_csv("https://drive.google.com/uc?export=download&id=1Hl2vfA3wf86wI5js9n6r9CMrteEh9ajo")
```

## Methods
* pd.concat() -> vertically stack 2 datasets
    - if you want to merge it horizontally use parameter axis = 1
* pd.append() -> vertically stack 2 datasets -> depricated
* joins -> to merge two or more dataframes on a common table
    - inner join
    - left join
    - right join
    - outer join
    - Syntax .merge(right table, how = "left/right/outer/inner", on = "column_name")

```python
# we have nov and dec registration data and we can easily merge them.
pd.concat([nov,dec])
# Here you can see we haveredundant index
regs = pd.concat([nov,dec],ignore_index = True)
# Whatif we need to retain it and also need to differenciate
pd.concat([nov,dec],keys=['nov','dec']) #multilevel indexing
```

```python
# inner join -> join
# items present in both the tables.
students.merge(regs,how = "inner",on="student_id")
```

```python
# left join
# include everything from left col
courses.merge(regs,how = "left",on = "course_id")
```

```python
# right join
# take everythhing from right table and join it with left col
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
```

```python
students.merge(regs,how = "right",on = "student_id")
```

```python
# outer join
students.merge(regs,how = "outer",on = "student_id")
```

## PRACTICE

```python
# 1. find total revenue generated
regs.merge(courses,how = "inner",on = "course_id")['price'].sum()
```

```python
# 2. find month by month revenue
temp = pd.concat([nov,dec],keys= ['nov','dec']).reset_index()
temp.merge(courses,how = "inner",on = "course_id").groupby('level_0').price.sum()
```

```python
# 3. Print the registration table
# cols -> name -> course -> price
regs.merge(students,on = "student_id").merge(courses, on = "course_id")[["name","course_name","price"]]
```

```python
# 4. Plot bar chart for revenue/course
regs.merge(courses,on = "course_id").groupby("course_name").price.sum().plot(kind= "bar")
```

```python
# 5. find students who enrolled in both the months
common_student_id = np.intersect1d(nov['student_id'],dec['student_id'])
common_student_id
```

```python
students[students['student_id'].isin(common_student_id)]
```

```python
# 6. find course that got no enrollment
# courses['course_id']
# regs['course_id']
course_id_list = np.setdiff1d(courses['course_id'],regs['course_id'])
courses[courses['course_id'].isin(course_id_list)]
```

```python
# 7. find students who did not enroll into any courses
student_id_list = np.setdiff1d(students['student_id'],regs['student_id'])
students[students['student_id'].isin(student_id_list)]
```

```python
# 8. Print student name -> partner name for all enrolled students
# self join
students.merge(students,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```

```python
# 9. find top 3 students who did most number enrollments
students.merge(regs,on = "student_id").groupby("name").course_id.nunique().sort_values(ascending = False).iloc[:3]
```

```python
# 10. find top 3 students who spent most amount of money on courses
regs.merge(students,on='student_id').merge(courses,on='course_id').groupby(['student_id','name'])['price'].sum().sort_values(ascending=False).head(3)
```
