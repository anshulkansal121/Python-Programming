## Example 1
5 features -> 1 neuron -> 1 output


```python
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    # Layer
    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self,features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

```python
# Create Dataset
features = torch.rand(10,5)

# Create an instance of model class
model = Model(features.shape[1])


# calling forward pass
model(features)
```

```python
# to check weights
model.linear.weight

# to check bias
model.linear.bias
```

## Example 2
we have 5 features -> hidden layer where we implement ReLu -> 1 Output Neuron using Sigmoid -> y pred

```python
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    # Layer
    self.linear1 = nn.Linear(num_features,3)
    self.linear2 = nn.Linear(3,1)
    self.ReLu = nn.ReLU()
    self.sigmoid = nn.Sigmoid()

  def forward(self,features):
    out = self.linear1(features)
    out = self.ReLu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
    return out
```

```python
# Create Dataset
features = torch.rand(10,5)

# Crete a Model instance
model = Model(features.shape[1])

# train / forward Pass
model(features)
```

```python
# this forward pass is going to be cubersome when we have multiple layers,
# solution - >Create a Sequential Container and use it in forward pass
```

```python
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    # Layer
    self.network = nn.Sequential(
    nn.Linear(num_features,3),
    nn.ReLU(),
    nn.Linear(3,1),
    nn.Sigmoid()
    )


  def forward(self,features):
    out = self.network(features)
    return out
```

```python
# Create Dataset
features = torch.rand(10,5)

# Crete a Model instance
model = Model(features.shape[1])

# train / forward Pass
model(features)
```

## Real-World Data

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

```python
df.shape
```

### Test - Train Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
y_train
```

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor = X_train_tensor.to(torch.float32)
```

```python
y_train_tensor = y_train_tensor.to(torch.float32)
```

```python
X_test_tensor = X_test_tensor.to(torch.float32)
y_test_tensor = y_test_tensor.to(torch.float32)
```

### Defining Model

```python
class Model(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.linear = nn.Linear(num_features,1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

```python
learning_rate = 0.1
epochs = 25
loss_function = nn.BCELoss()
```

```python
# Create an instance of Model Class
model = Model(X_train_tensor.shape[1])
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)
# defining loop
for epoch in range(epochs):
  # Forward pass
  ypred = model(X_train_tensor)

  # Calculate Loss
  loss = loss_function(ypred, y_train_tensor.view(-1,1))

  optimizer.zero_grad()
  # bakward pass
  loss.backward()

  # optimize / Update parameters
  optimizer.step()


```

```python
with torch.no_grad():
  ypred = model.forward(X_test_tensor)
  ypred = (ypred > 0.9).float()
  accuracy = (ypred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
