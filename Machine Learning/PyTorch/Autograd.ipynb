```python
import torch
```

```python
# require_grad =. True tells torch that we will be computing the Differention of this tensor and it automatically captures the operation
# performed on this tensor and returns the differentiation when needed
x = torch.tensor(3, requires_grad=True,dtype = torch.float32)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
y.backward()
```

```python
x.grad
```
