## Mathematical Operations

### 1. Scaler Operation

```python
x = torch.rand(2,2)
x
```

```python
# Scaler Operations
# addition
x+2
# substraction
x-2
# Multiplication
x*2
# Division
x/3
# int division
(x*100)//3
# mod
x%2
# power
x**2
```

### 2. Element-Wise Operation

```python
a = torch.rand(2,3)
b = torch.rand(2,3)

print(a)
print(b)
```

```python
# addition
a+b
```

```python
# subtraction
a-b
```

```python
# Division
a/b
```

```python
# Multiplication
a*b
```

```python
# Power
a**b
```

```python
c = torch.tensor([-1,2,-4,5,-6])
c
```

```python
# abs
torch.abs(c)
```

```python
# negative()
c.negative()
```

```python
# clamp
torch.clamp(c,min = 2, max= 9)
```

### 3. Reduction Operation

```python
t = torch.randint(size = (2,3), low = 2,high = 5,dtype = torch.float32)
t
```

```python
# sum
torch.sum(t)
```

```python
# column wise sum
torch.sum(t, dim = 0)
# row wise sum
torch.sum(t,dim = 1)
```

```python
# mean
torch.mean(t)
```

```python
# column wise mean
torch.mean(t,dim = 0)
```

```python
# median
torch.median(t)
```

```python
# max and min
torch.max(t)
torch.min(t)
```

```python
# std
torch.std(t)
```

```python
# argmin()
torch.argmin(t)
```

```python
# argmax()
torch.argmax(t)
```

### 4. Matrix Operations

```python
a = torch.randint(size = (2,3), low = 0, high = 10)
b = torch.randint(size = (3,2), low = 0, high = 10)

print(a)
print(b)
```

```python
# Matrix Multiplication
torch.matmul(a,b)
```

```python
# transpose
torch.transpose(a,0,1)
```

```python
# det
torch.det(a)
```

```python
# inverse()
torch.inverse(a)
```

### 5. Comparison Operations

```python
a = torch.randint(size = (2,3), low = 0, high = 10)
b = torch.randint(size = (2,3), low = 0, high = 10)
print(a)
print(b)
```

```python
# This is also element wise
# greater than
a > b

# less than
a < b

```

### 6. Special Functions

```python
k = torch.randint(size = (2,3), low = 0, high = 10,dtype = torch.float32)
k
```

```python
# log
torch.log(k)
```

```python
# exp
torch.exp(k)
```

```python
torch.sqrt(k)
```

```python
torch.sigmoid(k)
```

```python
torch.softmax(k,dim=0)
```

```python
torch.relu(k)
```
