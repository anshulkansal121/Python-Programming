# Project: Credit Risk Multi-Class Classification

## 1. Problem Statement
The objective of this project is to build a predictive model that classifies loan applicants into one of four risk categories—**P1, P2, P3, and P4**—based on historical credit bureau data and demographic information. 

These categories serve as priority levels that determine a customer's loan eligibility and interest rate pricing:
* **P1 & P2:** High-quality applicants with low to moderate risk.
* **P3:** Higher-risk applicants who may require stricter terms.
* **P4:** Very high-risk applicants who are likely to be ineligible for credit.

## 2. Technical Scope
The model utilizes a comprehensive dataset containing over 60 variables, categorized as follows:
* **Tradeline Information:** Total accounts, types of loans (Secured/Unsecured), and account ages.
* **Delinquency Metrics:** Historical "Days Past Due" (DPD) performance and frequency of missed payments.
* **Enquiry Behavior:** Frequency of credit searches in the last 3, 6, and 12 months.
* **Demographics:** Income, age, education, and employment stability.

## 3. Business Goal
By automating this classification, the institution aims to:
1.  **Reduce Credit Loss:** Identify and filter out high-risk (P4) individuals.
2.  **Optimize Approval Rates:** Streamline the process for low-risk (P1/P2) applicants.
3.  **Risk-Based Pricing:** Tailor interest rates based on the specific risk tier (P1–P4).

## 4. Evaluation Metrics
Success will be measured by the model's ability to accurately differentiate between the classes, prioritizing high **Recall for P4** (to avoid lending to risky customers) and high **Precision for P1** (to ensure premium customers are correctly identified).

## Here we cannot impute the missing values as the column hold different significance,
## We will drop all the columns that have more than 10000 rows of null value
## Rest we will the rows itself


```python
columns_to_be_removed = []
for i in df2.columns:
    if df2[df2[i]==-99999].shape[0] > 10000:
        columns_to_be_removed.append(i)
columns_to_be_removed
```

```python
df2.drop(columns_to_be_removed,axis=1,inplace = True)
```

```python
df2.shape
```

```python
for i in df2.columns:
    df2 = df2.loc[ df2[i] != -99999 ]
```

## Now both the datasets are cleaned, let's merge both the datasets
